```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,3,5]
plt.plot(x,y,color='green',linestyle='--',marker='o') #FOR PLOTING
plt.title("Example graph") #TITLE
plt.xlabel("x axis") #LABEL AXIS
plt.ylabel("y axis") #LABEL AXIS
plt.grid(True) #GRID ONLY LIST NOT ALLWED
plt.show() #SHOW GRAPH
```

```python
x=['A','B','C','D','E']
y=[8,4,6,3,5]
plt.bar(x,y,color='red')
plt.title("bar chart")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[8,4,6,3,5]
plt.barh(x,y,color='blue')
plt.title("bar chart horizontal")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,3,5]
plt.scatter(x,y,color='red')
plt.title("scatter graph")
plt.show()
```

```python
import numpy as np
data=np.random.randn(1000)
plt.hist(data,color="blue",bins=40)
plt.title("histogram")
plt.show()
```

```python
labels=["python","c++","java"]
sizes=[30,20,25]
plt.pie(sizes,labels=labels,autopct='%1.0f%%',startangle=90)
plt.title("pie chart")
plt.show()
```
